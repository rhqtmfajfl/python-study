# Colab에서 이미지 띄워주기

##matplotlib 사용하기

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show() # 여기서 비쥬얼 BGR 방식으로 사진을 불러온다.
```

```python
#위에서 기존 이미지의 색깔과 색이 바뀜
# open cv에서는
```

OpenCV는 RGB 채널 필터가 아닌 **BGR**필터로 사용하기 때문에 색상이 일반 RGB와는 다르다.

옛날 디지털 카메라의 처리 방식이 **BGR**이었기 때문이다!

### ⭐ cvtColor : 이미지 채널 변경 ⭐


```python
# cvtColor : 컬러 채널을 바꿔주기

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # 처음에 이미지가 들어감 그 다음에는 다양한 cv2 cvtcolor다양한 옵션들이 들어간다.
# BGR 채널을 RGB로 변환

plt.imshow(rgb_img)
plt.show()
```

```python
#gray scale 적용하기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, 'gray')
plt.show()
```
