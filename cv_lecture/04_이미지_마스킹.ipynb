# Load Image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
file_name = 'car1.jpg' #사진이름 정의
img = cv2.imread(file_name)
cv2_imshow(img)
```

```python
print(f'Height : {img.shape[0]} pixels')
print(f'Width : {img.shape[1]} pixels')
print(f'Channels : {img.shape[2]} Channels')
```

```python
# 중심좌표 구하기

(height, width) = img.shape[:2]
center = (width //2, height //2)

print('center cord : {}'.format(center))
```

# 이미지 마스킹

* 내가 보고싶은 부분만 보게하는 기법

```python
mask = np.zeros(img.shape[:2], dtype=np.uint8) # uint8 8비트로 채운다. zero를
cv2_imshow(mask)
```

```python
#위의검은 바탕에 원을 그림

mask_circle = cv2.circle(mask, center, int(height /2), (255,255,255), -1) # 그림을 그린 센터 좌표 255,255,255 새가지 색깔이 255가 되면 흰색이 된다.
#mask 배경화면  (파란, 초록 , 빨간) 세가지 색을 다 합치면 흰색
# int(height /2) 반지름
cv2_imshow(mask_circle)
```

```python
# 여러 비트 연산 중 and(&) 
#원본 이미지에 이미지 사이에 mak를 씌운다.
# mask 는 내가 원하는 부분 만보는것
# 필터링은  이미지 전체에 효과를 주는 것을 의미 한다.
# 윤곽선 따는 것 필터링이다.
```

```python
# 여러 비트 연산 중 and(&) 연산을 활용하면?
masked_bitwise = cv2.bitwise_and(img, img, mask=mask)
cv2_imshow(masked_bitwise)
```
